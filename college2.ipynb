```python
import pandas as pd
import numpy as np
```

```python
college = pd.read_csv('./data/college.csv')
college.head()
```

```python
pd.read_csv('./data/college_data_dictionary.csv')
```

```python
college[['INSTNM', 'STABBR', 'RELAFFIL', 'UGDS']]
```

```python
(
college[['INSTNM', 'STABBR', 'RELAFFIL', 'UGDS']]
    .groupby('STABBR', as_index=False)
    .apply(lambda df: df.sort_values(
    'UGDS', ascending=False).head(1)
    )
.droplevel(0)
.sort_values('UGDS', ascending=False)
)
```

```python
(
college[['INSTNM', 'STABBR', 'RELAFFIL', 'UGDS']]
    .groupby('RELAFFIL', as_index=False)
    .apply(lambda df: df.sort_values(
    'UGDS', ascending=False).head(1)
    )
.droplevel(0)
.sort_values('UGDS', ascending=False)
)
```
