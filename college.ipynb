```python
import pandas as pd
import numpy as np
```

```python
college = pd.read_csv('./data/college.csv')
# college.head()
college.info()
```

```python
# nlargest
```

```python
college['UGDS']
```

```python
college.shape
```

```python
college[['INSTNM', 'CITY', 'UGDS']].nlargest(7535, 'UGDS')
```

```python
college[['INSTNM', 'CITY', 'UGDS']].nlargest(7535, 'UGDS').drop_duplicates(subset='CITY')
```

```python
# groupby
```

```python
(
college[['INSTNM', 'CITY', 'UGDS']]
    .groupby('CITY', as_index=False)
    .apply(lambda df: df.sort_values(
        'UGDS', ascending=False).head(1)
          )
.droplevel(0)
.sort_values('UGDS', ascending=False)
)
```
